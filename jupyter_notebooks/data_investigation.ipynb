* To better visualise the dataset, decided to create a "age_group" column. Marketing-focused segments (from Claude.AI):

* **18-24**: Gen Z 
* **25-34**: Millennials
* **35-44**: Gen X Early
* **45-54**: Gen X Late
* **55-64**: Pre-Retirement
* **65-70**: Seniors

```python
# Define the age groups with the corresponding category label Â Â 
age_bins = [17, 24, 34, 44, 54, 64, 70]
age_group = ['Gen Z', 'Millennials', 'Gen X Early', 'Gen X Late', 'Pre-Retirement', 'Seniors']

# Find the position of the 'age' column Â Â 
age_index = df_cleaned.columns.get_loc('age')

# Use 'insert()' to place 'age_group' after 'age' columnÂ 
df_cleaned.insert(loc=age_index + 1, column='age_group', value=pd.cut(df_cleaned['age'], bins=age_bins, labels=age_group, right=True))

# pd.cut() converts continuous numerical data into categorical data by dividing it into discrete intervals (bins). Â 
# df_cleaned['age']: the continuous data (age column) to be categorised.Â 
# bins=[17, 24, 34, 44, 54, 64, 70]: the boundary values that define where to "cut" the data. Â 
# labels=['Gen Z', 'Millennials', ...]: custom names for each interval.Â 
# right=True: controls whether intervals include the right boundary.Â 

print(df_cleaned.head())
```
